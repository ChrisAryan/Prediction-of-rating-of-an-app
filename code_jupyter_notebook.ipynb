```python
rev1['Rating'] = rev1['Rating'].fillna(rev1['Rating'].mode()[0])
rev1.dropna(inplace=True)
rev1.isnull().sum()
```

```python
print(rev1.Size.unique())
```

```python
def mod_size(size):
    if size[-1]=="M":
        x=size[:-1]
        x=float(x)*1024
        return x
    elif size[-1]=="k":
        x=size[:-1]
        x=float(x)
        return x
    else:
        return 17000
    
rev1["Size"]=rev1["Size"].map(mod_size)
rev1.isna().sum()
```

```python
rev1["Reviews"]=rev1["Reviews"].astype(int)
```

```python
def mod_inst(install):
    x=install[:-1]
    x=x.replace(",","")
    x=int(x)
    return x

rev1["Installs"]=rev1["Installs"].map(mod_inst)
```

```python
rev1.Type.unique()
```

```python
def mod_type(type):
    if type=="Paid":
        return int(1)
    else:
        return int(0)
rev1["Type"]=rev1["Type"].map(mod_type)
rev1.Type.unique()
```

```python
rev1.Price.unique()
```

```python
def mod_price(price):
    if price=="0":
        return 0
    else:
        x=price[1:]
        x=float(x)
        return x

rev1["Price"]=rev1['Price'].map(mod_price)
rev1.Price.unique()
```
